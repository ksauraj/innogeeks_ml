```python
df
df.describe()
df[df['who'] == 'man'].survived.groupby(df.survived).count().plot(kind='pie', figsize=(6,6), explode=[0,0.1],autopct='%1.1f%%')
```

```python
df[df['pclass'] == 3].survived.groupby(df.survived).count().plot(kind='pie', figsize=(6,6), explode=[0,0.1],autopct='%1.1f%%', title="Survivial of 3rd class...")
```

```python
df[df['pclass'] == 1].survived.groupby(df.survived).count().plot(kind='pie', figsize=(6,6), explode=[0,0.1],autopct='%1.1f%%', title="Survivial of 1st class...")
```

```python
df[df['pclass'] == 2].survived.groupby(df.survived).count().plot(kind='pie', figsize=(6,6), explode=[0,0.1],autopct='%1.1f%%', title="Survivial of 2nd class...")
plt.axis("equal")
plt.legend(["Perished", "Survived"])
```

```python
age_surv=df[(df["age"] > 0) & (df["survived"]==1)]
age_died=df[(df["age"] > 0) & (df["survived"]==0)]
sns.distplot(age_surv["age"], color='g')
sns.kdeplot(age_died["age"], color='r')
sns.distplot(age_died["age"], color='r')
sns.kdeplot(age_surv["age"], color='g')
plt.title("HAHHAHHAHAH")
plt.xlable("Age", fontsize=15)
plt.ylable("Bacha ya Marr gaya", fontsize=15)
```

```python
survived_sex_counts = df.groupby(["survived", "who"]).size().unstack()
survived_sex_counts.plot(kind='bar', stacked=True)
survived_sex_counts.plot(kind='pie', stacked=True, subplots=True)
```
